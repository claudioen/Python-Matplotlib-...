WORKING ON PLOTS WITH UNCERTAINTIES

```python
#Load numpy for math/array operations
#and matplotlib for plotting
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Set up figure size and DPI for screen demo
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150
```

```python
nums = np.arange(0,10,0.5)
xerrors = np.random.uniform(size=20)
yerrors = np.random.uniform(size=20)
```

```python
#Basic errorbar plot
plt.errorbar(nums, np.sin(nums), xerr=xerrors, yerr=yerrors, ecolor='red')
```

```python
#Errorbar Line width
plt.errorbar(nums, np.sin(nums), xerr=xerrors, yerr=yerrors, elinewidth=0.1)
```

```python
#Errorbar cap size
plt.errorbar(nums, np.sin(nums), xerr=xerrors, yerr=yerrors, capsize=0)
```

```python
plt.bar(nums, np.sin(nums)**2, width=0.2)
```

```python
plt.bar(nums, np.sin(nums)**2, width=0.2, xerr=xerrors, yerr=yerrors)
```
