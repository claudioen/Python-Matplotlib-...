```python
# Load numpy for math/array operations
# and matplotlib for plotting
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Basic Line Plot
xvals = np.arange(0,10,0.1)
plt.plot(xvals, np.sin(xvals))
```

```python
# Axis Labels
plt.plot(xvals, np.sin(xvals))
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
# Multiple Plots
plt.plot(xvals, np.sin(xvals))
plt.plot(xvals, np.cos(xvals))
```

```python
# Legend & Title
plt.plot(xvals, np.sin(xvals), label=r"$y = \sin(x)$")
plt.plot(xvals, np.cos(xvals), label=r"$y = \cos(x)$")
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.legend()
```

```python
# Line Color
plt.plot(xvals, np.sin(xvals), label=r"$y = \sin(x)$", color='red', linestyle='--')
plt.plot(xvals, np.cos(xvals), label=r"$y = \cos(x)$", marker='o', linestyle='none')
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.legend()
```

```python
# Axis Limits -- Only positive values
plt.plot(xvals, np.sin(xvals), 'r--', label=r"$y = \sin(x)$")
plt.plot(xvals, np.cos(xvals), 'bo', label=r"$y = \cos(x)$")
plt.ylim(0,1)
plt.xlim(0,1)
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.legend()
```

```python
#Generate 1000 random numbers
rands = np.random.normal(size=1000)
```

```python
# Histogram
plt.hist(rands)
plt.xlabel(r'$x$')
plt.ylabel(r'$Pr(x)$')
```

```python
#Generate a smoothed, guassian random field
from scipy.ndimage.filters import gaussian_filter
rands2d = gaussian_filter(np.random.normal(size=(512,512)), sigma=10)
print(rands2d.shape)
```

```python
#show Image
plt.imshow(rands2d, cmap='magma')
plt.colorbar()
```
