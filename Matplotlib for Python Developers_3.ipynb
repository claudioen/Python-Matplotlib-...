```python
#Load numpy for math/array operations
#and matplotlib for plotting
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Set up figure size and DPI for screen demo
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150
```

```python
#basic bar plot
nums = np.random.uniform(size=10)
plt.bar(np.arange(10), height=nums)
```

```python
#horizontal bar plot
plt.barh(np.arange(10), width=nums)
```

```python
#bar color
plt.bar(np.arange(10), nums, color='limegreen')
```

```python
#bar alignment
plt.bar(np.arange(10), nums, align='center')
```

```python
#hatch & fill
plt.bar(np.arange(10), nums, hatch='-', color='none' )
```

```python
#bar width
plt.bar(np.arange(10), nums, width=0.5)
```

```python
#edge color
plt.bar(np.arange(10), nums, width=1, edgecolor= 'k', color='g')
```

```python
#bottom alignment
plt.bar(np.arange(10), nums, bottom=nums)
```

```python
#basic histogram
rands = np.random.normal(size=int(1e6))
plt.hist(rands)
```

```python
#return type
print(plt.hist(rands))
plt.clf()
```

```python
#range
plt.hist(rands, range=(0,1))
```

```python
#bin choice
plt.hist(rands,bins=[0,1,2])
```

```python
#histtype
plt.hist(rands, bins='auto')
```

```python
#normalization
plt.hist(rands, bins='auto', histtype='stepfilled')
```

```python
#cumulative histogram
plt.hist(rands, bins='auto')
```

```python
#cumulative histogram
plt.hist(rands, bins='auto', cumulative=-1, normed=True)
```

```python
#multiple histograms
#plt.hist(rands, bins='auto', histtype='stepfilled')
plt.hist((rands, 0.5*rands), bins='auto', histtype='stepfilled', staked=True)
#plt.hist((rands, 0.5*rands), bins='auto', histtype='stepfilled'
```
