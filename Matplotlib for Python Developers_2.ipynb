DIFFERENTIATING LINE AND SCATTER PLOTS

```python
#Load numpy for math/RRy operations
#and matplotlib for plotting
import numpy  as np
import matplotlib.pyplot as plt
```

```python
#Set up figure sizer and DPI for screen demo
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['figure.dpi'] = 150
```

```python
#Basic Line plot
xvals = np.arange(0,10,0.1)
plt.plot(xvals, np.sin(xvals))
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
#different line style
idx = 0
for marker in ('-', '--', ':', '-.'):
    plt.plot(xvals, np.sin(xvals)+idx, ''.join(('k', marker)))
    idx += 1
```

```python
#Markers
plt.plot(xvals, np.sin(xvals), 'k.')
plt.xlabel(r'$x$')
plt.ylabel(r'\sin(x)$')
```

```python
#Marker frequency (markevery)
xvals2 = np.arange(0,10,0.01)
plt.plot(xvals2, np.sin(xvals2), 'k.')
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
#Marker frequency (markevery)
xvals2 = np.arange(0,10,0.01)
plt.plot(xvals2, np.sin(xvals2), 'k.', markevery=10)
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
#Marker size
plt.plot(xvals, np.sin(xvals), 'r.', markersize=40)
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
#Different markers
idx = 0
for marker in ('.', ',', 'o', 'v', '^', '>', '<', '1', '2', '3', '4', 's', 'p', '*', 'h', 'H', '+', 'x', 'D', '|', '_'):
               plt.plot(idx, 0, ''.join(('r', marker)))
               idx += 1
plt.xlim(-0.5, idx+0.5)
```

```python
#Latex marker
plt.plot(xvals, np.sin(xvals), color='k', linestyle='none', marker=r'$\sin$')
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
#line width
plt.plot(xvals, np.sin(xvals), linewidth=1)
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
#zorder
plt.plot(xvals, np.sin(xvals), lw=8, zorder=10)
plt.plot(xvals+1, np.sin(xvals), lw=8, zorder=1)
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
#transparency
plt.plot(xvals, np.sin(xvals), lw=8, alpha=0.5)
plt.plot(xvals+1, np.sin(xvals), lw=8, alpha=0.5)
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
#array order matters!
shuffled_xvals = np.random.permutation(xvals)
plt.plot(shuffled_xvals, np.sin(shuffled_xvals), 'k.')
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
#basic scatter plot
xvals = np.arange(0,10,0.1)
plt.scatter(xvals, np.sin(xvals))
plt.xlabel(r'$x$')
plt.ylabel(r'\sin(x)$')
```

```python
#colours
plt.scatter(xvals, np.sin(xvals), c=np.cos(xvals))
plt.colorbar()
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
#colour maps
plt.scatter(xvals, np.sin(xvals), c=np.cos(xvals), cmap='inferno')
plt.colorbar()
plt.xlabel(r'$x$')
plt.ylabel((r'$\sin(x)$'))
```

```python
#marker edge colors
plt.scatter(xvals, np.sin(xvals), s=1, c=np.cos(xvals), edgecolor='none', marker=',', )
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
#sizes
plt.scatter(xvals, np.sin(xvals), c=np.cos(xvals), s=np.power(xvals, 3))
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```
